# Automating Opinium calculations

```python
import pandas as pd
```

```python
#Import Opinium data into a pandas df.
#I have previously formatted it so that it has unique headers + brand name and weighting type in each row.
#...then downloaded it as csv.

df = pd.read_csv('data.csv')
df.head()
```

```python
#Get weighted data only
dfw = df.loc[df['Weighting']=='weighted']
```

```python
dfw.head()
```

```python
#Import brand names that are in the CX wave
CXbrandlist = pd.read_csv('brndlst.csv', header=None)
brndlst = list(CXbrandlist[0].values)
brndlst
```

```python
#Select data for these brands only for easier processing.
dfw2 = dfw.loc[dfw['Brand'].isin(brndlst)]
```

```python
#Remove the weird Opinium *s and setting brand name as index.
dfw2.replace('*', 0, inplace=True)
dfw2.set_index('Brand', inplace=True)
dfw2.head()
```

```python
#Remove trailing whitespaces from column names (only there because of the way I replaced things in the names, but maybe a good safety feature anyway)
dfw2 = dfw2.rename(columns=lambda x: x.strip())
dfw2.columns
```

```python
#Define a function that gets a response on the basis of brand, base, option, dataframe.

def getresponse(brand, base, option, df):
    return float(df.loc[brand][f'{base} {option}'])

#e.g., getresponse('4paws', 'Male', 'Extremely', dfw2)
```

```python
#Generate calculations.
def generate_calcs(dataframe, bases, options, calc):
    #Define output dict.
    happiness_dict=dict()
    for brand in dataframe.index:
        happiness_dict[brand]=dict()
        for base in bases:
            h = list()
            for option in options:
                h.append(getresponse(brand, base, option, dataframe))
            happiness_dict[brand][f'{base} happiness']=calc(h)
    return happiness_dict
```

```python
#Define bases and options.
bases = ['All', 'Male', 'Female', '18-30', '31-45', '46-64', '65p', 'SNIW', 'North', 'Mid', 'London', 'South']
options = ['Extremely', 'Fairly', 'Neither', 'Fairly dis', 'Extremely dis', 'NET']

#Define a calculation.
def happiness_calc(h):
    return((h[0]*2+h[1]-h[3]-2*h[4])/(2*h[5])) #The happiness calculation.

#Generate happiness calculation.
happiness_dict = generate_calcs(dfw2, bases, options, happiness_calc)
```

```python
#Export results to csv.
out_df = pd.DataFrame.from_dict(happiness_dict, 'index')
out_df.head()
out_df.to_csv('out.csv')
```
